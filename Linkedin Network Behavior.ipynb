```python
df_clean=df[(df["companyStaffCount"]<20000)&
(df["conn"]<1000)]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

plt.scatter(df_clean["companyStaffCount"],
            df_clean["conn"],
            alpha=0.5)

plt.xlabel("Company Staff Count")
plt.ylabel("Linkedin Connections")
plt.title("Company Size vs Professional Network Strength")

plt.show()
```

```python
df_clean=df[(df["companyStaffCount"]<20000)&
(df["conn"]<1000)].copy()

df_clean["SizeCategory"]=pd.cut(
    df_clean["companyStaffCount"],
    bins=[0,50,200,1000,5000,20000],
    labels=["Small","Medium","Large","Enterprise","Global"]
)
```

```python
df_clean.shape
```

```python
df_clean["SizeCategory"].value_counts()
```

```python
df_clean=df.copy()
df_clean=df_clean.dropna(subset=["companyStaffCount",
"conn"])
```

```python
df_clean["SizeCategory"]=pd.cut(
    df_clean["companyStaffCount"],
    bins=5
)
```

```python
df_clean["SizeCategory"].value_counts()
```

```python
avg_conn=df_clean.groupby("SizeCategory")["conn"].mean()

import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
avg_conn.plot(kind="bar")

plt.xlabel("Company Size Category")
plt.ylabel("Average Connections")
plt.title("Average Connections vs Company Size")
plt.show()
```

```python
avg_conn=df_clean.groupby("SizeCategory")["conn"].mean()

import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

ax=avg_conn.plot(kind="bar")

plt.xlabel("Company Size Category")
plt.ylabel("Average Linkedin Connections")
plt.title("Average Linkedin Connections by Company Size")

# Add numbers on top of bars

for i, value in enumerate(avg_conn):
    plt.text(i,value,round(value,1),ha='center',va='bottom')

plt.show()
```
